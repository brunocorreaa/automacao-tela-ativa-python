```python
"""
Script de automação para manter a tela ativa.

Objetivo:
Evitar que o computador entre em modo de suspensão ou bloqueio
em ambientes onde dashboards em Power BI ficam expostos em telas
ou televisores para monitoramento contínuo.

O script move levemente o cursor do mouse em intervalos de tempo,
simulando atividade do usuário.
"""

import pywinauto.mouse as mouse
import time

# Coordenadas utilizadas para o movimento do mouse
POSICAO_1 = (90, 90)
POSICAO_2 = (100, 100)

# Intervalo entre movimentos (segundos)
INTERVALO = 100

print("Script iniciado. Movendo cursor periodicamente para evitar bloqueio da tela...")

while True:
    
    # Move o mouse para a primeira posição
    mouse.move(coords=POSICAO_1)
    
    # Aguarda o intervalo definido
    time.sleep(INTERVALO)
    
    # Move o mouse para a segunda posição
    mouse.move(coords=POSICAO_2)
    
    # Aguarda novamente
    time.sleep(INTERVALO)
```
